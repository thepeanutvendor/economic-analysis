```python
import numpy as np
import pandas
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize=(12.8, 6.4), facecolor='#2F3F3C')
ax.set_facecolor('#2F3F3C')

# Create random-looking but correlated data
np.random.seed(42)
x = np.random.normal(5, 1.5, 50)
y = -0.6*x + 8 + np.random.normal(0, 0.8, 50)

ax.scatter(x, y, color='#AEB6B2', s=80, alpha=0.6, edgecolors='none')
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax.plot(np.sort(x), p(np.sort(x)), color='#D6D1B8', linewidth=3)

# Clean axes
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.savefig('featured.png', dpi=150)
```
